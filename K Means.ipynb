```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
#load data set
univ=pd.read_csv('C:/Users/ASUS/universities.csv')
```

```python
univ
```

```python
#Normalization function
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaled_univ_df=scaler.fit_transform(univ.iloc[:,1:])
```

```python
wcss=[]
for i in range (1,11):
    kmeans= KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_univ_df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

```python
#Build cluster
cluster_new=KMeans(4,random_state=42)
cluster_new.fit(scaled_univ_df)
```

```python
cluster_new.labels_
```

```python
#Assign clusters to data set
univ['clusterid_new']=cluster_new.labels_
```

```python
#these are standerized value
cluster_new.cluster_centers_
```

```python
univ.groupby('clusterid_new').agg(['mean']).reset_index()
```
